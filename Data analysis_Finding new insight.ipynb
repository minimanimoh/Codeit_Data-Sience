```python
brand_country = {
    'Dell': 'U.S.', 
    'Apple': 'U.S.',
    'Acer': 'Taiwan',
    'HP': 'U.S.',
    'Lenovo': 'U.S.', 
    'Alienware': 'China', 
    'Microsoft': 'U.S.',
    'Asus': 'U.S.'
}
```

```python
laptops['brand'].map(brand_country)
```

```python
laptops['brand_country'] = laptops['brand'].map(brand_country)
```

```python
laptops
```

```python
country_groups = laptops.groupby('brand_country')
```

```python
price_df = pd.read_csv('downloads/vegetable_price.csv')
```

```python
quan_df= pd.read_csv('downloads/vegetable_quantity.csv')
```

```python
price_df
```

```python
quan_df
```

```python
#inner join
pd.merge(price_df, quan_df, on='Product')
```

```python
#left join
pd.merge(price_df, quan_df, on='Product', how='left')
```

```python
#right join
pd.merge(price_df, quan_df, on='Product', how='right')
```

```python
#outer join
pd.merge(price_df, quan_df, on='Product', how='outer')
```
